```python
OUTPUT_FILE = "/content/drive/MyDrive/Edulift/eVED_171101_week_clean.csv"
```

```python
df = df.rename(columns={
        "Timestamp(ms)": "timestamp_ms",
        "Latitude[deg]": "lat",
        "Longitude[deg]": "lon",
        "Vehicle Speed[km/h]": "speed_kmh",
        "Elevation Raw[m]": "elev_raw_m",
        "Elevation Smoothed[m]": "elev_smooth_m"
})
```

```python
 df["timestamp"] = pd.to_datetime(df["timestamp_ms"], unit="ms", errors="coerce")
```

```python
df = df.dropna(subset=["lat", "lon"])
```

```python
df = df[(df["lat"].between(-90, 90)) & (df["lon"].between(-180, 180))]
```

```python
df = df[df["speed_kmh"].between(0, 200)]
```

```python
df = df.drop_duplicates(subset=["timestamp", "lat", "lon"])
```

```python
df.head()
```

```python
 keep_cols = ["VehId", "Trip", "timestamp_ms", "timestamp",
                 "lat", "lon", "speed_kmh", "elev_raw_m", "elev_smooth_m"]
```

```python
df = df[keep_cols]
```

```python
df.to_csv(OUTPUT_FILE, index=False)
```
